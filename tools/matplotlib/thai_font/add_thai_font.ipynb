###  ติดตั้ง Sarabun font ทั้งหมด

#### ดาวน์โหลด font ทั้งหมดใน GIT repository จากที่นี่
~~~bash
    wget https://go.wiphoo.dev/f6OtGJ -O 'NotoSans[wdth,wght].ttf'
    wget https://go.wiphoo.dev/vHsbYE -O 'NotoSans-Italic[wdth,wght].ttf'
    wget https://go.wiphoo.dev/D8Xrn2 -O 'NotoSansThai[wdth,wght].ttf'
~~~

#### ติดตั้ง font ทั้งหมด
~~~bash
    # Ubuntu
    mkdir -p ~/.fonts
    cp ./NotoSans*.ttf ~/.fonts/

    # MacOS
    cp ./NotoSans*.ttf ~/Library/Fonts/
~~~

```python
import matplotlib.font_manager as fm

# Ubuntu
font_dir = '~/.fonts'
# MacOS
# font_dir = '~/Library/Fonts'

# load all font files
font_files = fm.findSystemFonts(fontpaths=[font_dir])
for font_file in font_files:
    fm.fontManager.addfont(font_file)


import matplotlib.pyplot as plt
# set default font
plt.rcParams['font.family'] = ['Noto Sans Thai', 'Noto Sans']
```

```python
# Plot histogram keyword vs frequencies
keyword_counts.plot(kind='bar', color='skyblue', edgecolor='black')

# Add labels and title
plt.xlabel('Keywords', fontdict={"size":15, 'color':'grey', 'weight':'bold'})
plt.ylabel('Frequency')
plt.title('Histogram of Thai Food (อาหาร) Keyword Frequencies', fontdict={"size":17, 'color':'grey', 'style': 'italic', 'weight':'bold'})

# Show the plot
plt.show()
```
